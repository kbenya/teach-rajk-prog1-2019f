```python
counter = 0
you = []

current = ['YOU']

for i in range(len(objektum)):
    new_list = []
    for pair in total:
        if pair[1] in current:
            print("pair", pair[1])
            new_list.append(pair[0])
            you.append(pair[0])
            counter += 1
            current = new_list
print(you)
```

```python
counter = 0
san = []

current = ['SAN']

for i in range(len(objektum)):
    new_list = []
    for pair in total:
        if pair[1] in current:
            print("pair", pair[1])
            new_list.append(pair[0])
            san.append(pair[0])
            counter += 1
            current = new_list
print(san)
```

```python
solution2 = []
for g in you:
    if g not in san:
        solution2.append(g)
        
for h in san:
    if h not in you:
        solution2.append(h)
        
print(set(solution2))
print(len(set(solution2)))
        
```
